```python
import pandas as pd

# Biblioteca que faz chamada para o Google Drive
from google.colab import drive

# Isso solicitará autorização com a sua conta do Google Drive
drive.mount('/content/drive')

# Cria a variável data e lê o arquivo csv, index_col = 0 retira a coluna de índices, parse_dates = true diz quais colunas devem ser interpretadas como data
data = pd.read_csv("/content/drive/My Drive/Notebooks/AAPL.csv", index_col=0, parse_dates=True)

# Mostra as primeiras linhas do dataset
data.head()

```

```python
# Mostra a diferença do valor entre o fechamento e início por data
daily_change = data['Close'] - data['Open']
daily_change
```

```python
# Percentual da diferença do valor entre o fechamento e o início por data
daily_change_percent = data['Close'] - data['Open'] / data['Open'] * 100
daily_change_percent
```

```python
# Mostra o crescimento do fechamento por data a partir da data inicial
norm = data['Close']/data['Close'].iloc[0]
norm
```

```python
# Adicionando novas colunas com os calculos 
data ['daily_change'] = data['Close'] - data['Open']
data ['daily_change_percent'] = data['Close'] - data['Open'] / data['Open'] * 100
data ['norm'] = data['Close']/data['Close'].iloc[0]
data.head()
```
